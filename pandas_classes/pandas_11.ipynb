Reading/Writing Data to Different Sources - Excel, JSON, SQL

```python
import pandas as pd
```

```python
# Working with CSV file
na_vals = ['NA', 'Missing', ' ']
df = pd.read_csv('Data/stack-overflow-developer-survey-2019/survey_results_public.csv', index_col='Respondent', na_values=na_vals)
schema_df = pd.read_csv('Data/stack-overflow-developer-survey-2019/survey_results_schema.csv',  index_col='Column')
# missing values are defined while reading the CSV
```

```python
filt = (df['Country'] == 'Germany')
germany_df = df.loc[filt]
germany_df
```

```python
# Let's write the survey from Germany to file
germany_df.to_csv('Data/analyzed/stack-overflow-developer-survey-2019-Germany.csv')
```

```python
# Let's work with tab delimited file
germany_df.to_csv('Data/analyzed/stack-overflow-developer-survey-2019-Germany.tsv', sep='\t')
```

```python
# Working with excel
# Let's write the survey from Germany to file
germany_df.to_excel('Data/analyzed/stack-overflow-developer-survey-2019-Germany.xlsx')
```

```python
germany_xlsx = pd.read_excel('Data/analyzed/stack-overflow-developer-survey-2019-Germany.xlsx', index_col='Respondent')
germany_xlsx.head()
```

```python
# Let's work with JSON file
# By default it's dictionary like
germany_df.to_json('Data/analyzed/stack-overflow-developer-survey-2019-Germany.json')
```

```python
# list like json
germany_df.to_json('Data/analyzed/stack-overflow-developer-survey-2019-Germany1.json', orient='records', lines=True)
```

```python
germany_json = pd.read_json('Data/analyzed/stack-overflow-developer-survey-2019-Germany1.json', orient='records', lines=True)
germany_json.head(5)
```
